```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
import matplotlib.patches as mpatches

# Configuration du style des graphiques
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['legend.fontsize'] = 10

def load_and_prepare_data():
    """Charger et préparer les données des deux fichiers CSV"""
    try:
        # Charger les données originales
        #tuna_data = pd.read_csv('D:/2019-2020-2021/data/tuna_data_namzaf0122.csv', encoding='latin-1')
        tuna_data = pd.read_csv('D:/2019-2020-2021/data/tuna_data2.csv', encoding='latin-1')
        # Charger les données avec prédictions
        tuna_clean = pd.read_csv('D:/2019-2020-2021/data/tuna_data_clean.csv', encoding='latin-1')
        
        print(f"Données originales: {tuna_data.shape}")
        print(f"Données nettoyées: {tuna_clean.shape}")
        
        return tuna_data, tuna_clean
    except FileNotFoundError as e:
        print(f"Erreur: {e}")
        print("Assurez-vous que les fichiers 'tuna_data.csv' et 'tuna_data_clean.csv' sont dans le répertoire de travail.")
        return None, None

def aggregate_by_year(df, group_cols, agg_cols):
    """Agréger les données par année"""
    return df.groupby(group_cols)[agg_cols].sum().reset_index()

def calculate_correlation(x, y):
    """Calculer la corrélation de Pearson"""
    # Retirer les valeurs NaN
    mask = ~(np.isnan(x) | np.isnan(y))
    if np.sum(mask) > 1:
        corr, p_value = pearsonr(x[mask], y[mask])
        return corr, p_value
    return np.nan, np.nan

def create_comparison_plot_style(merged_data, catch_col, pred_col, title):
    """Créer un graphique dans le style de l'exemple fourni"""
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    years = merged_data['Year']
    observed = merged_data[catch_col]
    predicted = merged_data[pred_col]
    
    # Calculer la corrélation
    corr, p_value = calculate_correlation(observed.values, predicted.values)
    
    # Tracer les lignes avec marqueurs (style similaire à l'exemple)
    line1 = ax.plot(years, observed, 'o-', color='black', linewidth=2, 
                   markersize=6, markerfacecolor='black', label=f'Raw {title}')
    
    line2 = ax.plot(years, predicted, 'o-', color='red', linewidth=2, 
                   markersize=6, markerfacecolor='white', markeredgecolor='red',
                   markeredgewidth=2, linestyle='-', label=f'Standardized {title}')
    
    # Configuration du graphique
    ax.set_title(f'Comparison of Raw and Standardized {title}', fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel('Year', fontsize=14)
    ax.set_ylabel(f'{title}', fontsize=14)
    
    # Grille
    ax.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)
    ax.set_axisbelow(True)
    
    # Légende
    ax.legend(loc='upper right', frameon=True, fancybox=True, shadow=True)
    
    # Ajuster les marges
    ax.margins(x=0.02, y=0.05)
    
    # Configuration de l'axe Y pour qu'il commence à 0
    y_min = 0
    y_max = max(max(observed), max(predicted)) * 1.1
    ax.set_ylim(y_min, y_max)
    
    # Espacement des années sur l'axe X
    year_range = max(years) - min(years)
    if year_range > 15:
        ax.set_xticks(years[::2])  # Afficher une année sur deux si beaucoup d'années
    else:
        ax.set_xticks(years)
    
    plt.tight_layout()
    
    # Afficher les statistiques
    print(f"\nStatistiques pour {title}:")
    print(f"  Corrélation: {corr:.3f} (p-value: {p_value:.3f})")
    print(f"  R²: {corr**2:.3f}")
    print(f"  Moyenne observée: {observed.mean():.2f}")
    print(f"  Moyenne prédite: {predicted.mean():.2f}")
    
    plt.show()

def create_all_comparisons():
    """Créer tous les graphiques de comparaison"""
    # Charger les données
    tuna_data, tuna_clean = load_and_prepare_data()
    
    if tuna_data is None or tuna_clean is None:
        return
    
    # Agréger par année
    catch_yearly = aggregate_by_year(
        tuna_data, 
        ['Year'], 
        ['Catch_ALB', 'Catch_BET', 'Catch_SWO', 'Catch_YFT']
    )
    
    pred_yearly = aggregate_by_year(
        tuna_clean, 
        ['Year'], 
        ['glm_alb_pred', 'glm_bet_pred', 'glm_swo_pred', 'glm_yft_pred']
    )
    
    # Fusionner les données
    merged_data = pd.merge(catch_yearly, pred_yearly, on='Year', how='inner')
    
    # Note: Il y a une erreur dans votre demande initiale - vous avez écrit "glm_alb_yft" au lieu de "glm_yft_pred"
    # Je vais utiliser la correspondance correcte
    
    # Définir les comparaisons à effectuer
    comparisons = [
        ('Catch_ALB', 'glm_alb_pred', 'CPUE (Albacore)'),
        ('Catch_BET', 'glm_bet_pred', 'CPUE (Bigeye)'),
        ('Catch_SWO', 'glm_swo_pred', 'CPUE (Swordfish)'),
        ('Catch_YFT', 'glm_yft_pred', 'CPUE (Yellowfin)')
    ]
    
    print("Création des graphiques de comparaison...")
    print("="*60)
    
    # Créer un graphique pour chaque comparaison
    for catch_col, pred_col, title in comparisons:
        create_comparison_plot_style(merged_data, catch_col, pred_col, title)

def create_combined_plot():
    """Créer un graphique combiné avec tous les types de thon"""
    # Charger les données
    tuna_data, tuna_clean = load_and_prepare_data()
    
    if tuna_data is None or tuna_clean is None:
        return
    
    # Agréger par année
    catch_yearly = aggregate_by_year(
        tuna_data, 
        ['Year'], 
        ['Catch_ALB', 'Catch_BET', 'Catch_SWO', 'Catch_YFT']
    )
    
    pred_yearly = aggregate_by_year(
        tuna_clean, 
        ['Year'], 
        ['glm_alb_pred', 'glm_bet_pred', 'glm_swo_pred', 'glm_yft_pred']
    )
    
    # Fusionner les données
    merged_data = pd.merge(catch_yearly, pred_yearly, on='Year', how='inner')
    
    # Créer une figure avec 4 sous-graphiques
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Comparison of Raw and Standardized CPUE for Different large pelagic Species', 
                 fontsize=25, fontweight='bold')
    
    comparisons = [
        ('Catch_ALB', 'glm_alb_pred', 'Albacore (ALB)'),
        ('Catch_BET', 'glm_bet_pred', 'Bigeye (BET)'),
        ('Catch_SWO', 'glm_swo_pred', 'Swordfish (SWO)'),
        ('Catch_YFT', 'glm_yft_pred', 'Yellowfin (YFT)')
    ]
    
    axes = axes.flatten()
    
    for i, (catch_col, pred_col, title) in enumerate(comparisons):
        ax = axes[i]
        
        years = merged_data['Year']
        observed = merged_data[catch_col]
        predicted = merged_data[pred_col]
        
        # Tracer les lignes
        ax.plot(years, observed, 'o-', color='black', linewidth=2, 
               markersize=5, markerfacecolor='black', label=f'Raw CPUE')
        
        ax.plot(years, predicted, 'o-', color='red', linewidth=2, 
               markersize=5, markerfacecolor='white', markeredgecolor='red',
               markeredgewidth=1.5, label=f'Standardized CPUE')
        
        # Configuration
        ax.set_title(title, fontsize=18, fontweight='bold')
        ax.set_xlabel('Year',  fontsize=18)
        ax.set_ylabel('CPUE', fontsize=18)
        #ax.grid(True, alpha=0.3)
        ax.legend()
        # Espacement des années sur l'axe X
        year_range = max(years) - min(years)
        if year_range > 10:
           ax.set_xticks(years[::1])  # Afficher une année sur deux si beaucoup d'années
        else:
           ax.set_xticks(years, fontsize=18)
        # Calculer et afficher la corrélation
        corr, _ = calculate_correlation(observed.values, predicted.values)
        ax.text(0.02, 0.98, f'r = {corr:.3f}', transform=ax.transAxes, 
               verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        ax.tick_params(axis='x', rotation=45, labelsize=18)
        ax.tick_params(axis='y', labelsize=18)
    
    plt.tight_layout()
    plt.show()

def main():
    """Fonction principale"""
    print("Comparaison des données de capture de thon vs prédictions GLM")
    print("="*60)
    
    # Créer les graphiques individuels (style de l'exemple fourni)
    create_all_comparisons()
    
    # Optionnel: créer aussi un graphique combiné
    print("\nCréation du graphique combiné...")
    create_combined_plot()

if __name__ == "__main__":
    main()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Rectangle

# Charger les données
df = pd.read_csv('D:/2019-2020-2021/data/tuna_data_clean.csv')

# Configuration générale des graphiques
plt.style.use('default')
sns.set_palette("Set2")

# Dictionnaire pour convertir les numéros de mois en noms abrégés
month_names = {
    1: 'Jan.', 2: 'Feb.', 3: 'Mar.', 4: 'Apr.',
    5: 'May.', 6: 'Jun.', 7: 'Jul.', 8: 'Aug.',
    9: 'Sept.', 10: 'Oct.', 11: 'Nov.', 12: 'Dec.'
}

def get_month_labels(months):
    """Convertir les numéros de mois en noms abrégés"""
    return [month_names.get(month, str(month)) for month in months]

# =================== BOX PLOTS ===================
def create_boxplots():
    """Créer les box plots pour toutes les variables en fonction du mois"""
    
    # Variables à analyser
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    
    # Créer une figure avec subplots
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    #fig.suptitle('Box plots des variables océanographiques par mois', fontsize=16, fontweight='bold')
    
    # Aplatir les axes pour faciliter l'itération
    axes_flat = axes.flatten()
    
    for i, var in enumerate(variables):
        ax = axes_flat[i]
        
        # Créer le box plot
        box_data = []
        months = sorted(df['Month'].unique())
        
        for month in months:
            month_data = df[df['Month'] == month][var].dropna()
            box_data.append(month_data)
        
        # Box plot avec style personnalisé
        month_labels = get_month_labels(months)
        bp = ax.boxplot(box_data, labels=month_labels, patch_artist=True, 
                       widths=0.6, showfliers=True)
        
        # Personnaliser les couleurs
        colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink', 'lightgray']
        for j, patch in enumerate(bp['boxes']):
            patch.set_facecolor(colors[j % len(colors)])
            patch.set_alpha(0.7)
        
        # Personnaliser les outliers
        for flier in bp['fliers']:
            flier.set(marker='o', color='red', alpha=0.6, markersize=3)
        
        # Étiquettes et titre
        ax.set_xlabel('Month', fontweight='bold')
        ax.set_ylabel(f'{var}', fontweight='bold', fontsize =18)
        ax.set_title(f'{chr(65+i)}) {var}', fontweight='bold', loc='left', fontsize =18)
        ax.grid(True, alpha=0.3)
        
        # Rotation des étiquettes de mois pour une meilleure lisibilité
        ax.tick_params(axis='x', rotation=45, labelsize=18)
        ax.tick_params(axis='y', labelsize=18)
    
        # Ajuster les limites si nécessaire
        if var != 'SSH1':  # SSH peut avoir des valeurs négatives
            ax.set_ylim(bottom=0)
    
    plt.tight_layout()
    plt.savefig('boxplots_variables_mois.png', dpi=300, bbox_inches='tight')
    plt.show()

# =================== HISTOGRAMMES ===================
def create_histograms():
    """Créer les histogrammes pour toutes les variables"""
    
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    
    # Créer une figure avec subplots
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    #fig.suptitle('Histogrammes des variables océanographiques', fontsize=16, fontweight='bold')
    
    # Aplatir les axes pour faciliter l'itération
    axes_flat = axes.flatten()
    
    for i, var in enumerate(variables):
        ax = axes_flat[i]
        
        # Données pour l'histogramme
        data = df[var].dropna()
        
        # Créer l'histogramme
        n, bins, patches = ax.hist(data, bins=30, alpha=0.7, color='gray', 
                                  edgecolor='black', linewidth=0.5)
        
        # Étiquettes et titre
        ax.set_xlabel(f'{var}', fontweight='bold', fontsize=18)
        ax.set_ylabel('Observations No.', fontweight='bold', fontsize=18)
        ax.set_title(f'{chr(65+i)})', fontweight='bold', loc='left', fontsize=18)
        ax.grid(True, alpha=0.3)
        ax.tick_params(axis='x', labelsize=18)
        ax.tick_params(axis='y', labelsize=18)
        
        # Ajouter des statistiques
        mean_val = data.mean()
        std_val = data.std()
        #ax.axvline(mean_val, color='red', linestyle='--', alpha=0.7, 
                  #label=f'Moyenne: {mean_val:.2f}')
        
        # Ajouter la légende uniquement si il y a de la place
        #if i < 3:  # Première ligne
            #ax.legend(fontsize=8)
        
    plt.tight_layout()
    plt.savefig('histogrammes_variables.png', dpi=300, bbox_inches='tight')
    plt.show()

# =================== ANALYSE PAR MOIS SÉPARÉE ===================
def create_monthly_analysis():
    """Créer des analyses séparées par mois pour chaque variable"""
    
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    months = sorted(df['Month'].unique())
    
    for var in variables:
        fig, axes = plt.subplots(2, 1, figsize=(12, 10))
        
        # Box plot
        ax1 = axes[0]
        box_data = []
        for month in months:
            month_data = df[df['Month'] == month][var].dropna()
            box_data.append(month_data)
        
        month_labels = get_month_labels(months)
        bp = ax1.boxplot(box_data, labels=month_labels, 
                        patch_artist=True, widths=0.6)
        
        colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink', 'lightgray']
        for j, patch in enumerate(bp['boxes']):
            patch.set_facecolor(colors[j % len(colors)])
            patch.set_alpha(0.7)
        
        # Personnaliser les outliers
        for flier in bp['fliers']:
            flier.set(marker='o', color='red', alpha=0.6, markersize=3)
        
        ax1.set_title(f'Box plot - {var} par mois', fontweight='bold')
        ax1.set_ylabel(var, fontweight='bold')
        ax1.set_xlabel('Mois', fontweight='bold')
        ax1.grid(True, alpha=0.3)
        ax1.tick_params(axis='x', rotation=45)
        
        # Histogramme global
        ax2 = axes[1]
        data = df[var].dropna()
        n, bins, patches = ax2.hist(data, bins=30, alpha=0.7, color='gray', 
                                   edgecolor='black', linewidth=0.5)
        
        # Ajouter statistiques sur l'histogramme
        mean_val = data.mean()
        median_val = data.median()
        ax2.axvline(mean_val, color='red', linestyle='--', alpha=0.8, 
                   label=f'Moyenne: {mean_val:.2f}')
        ax2.axvline(median_val, color='blue', linestyle='--', alpha=0.8, 
                   label=f'Médiane: {median_val:.2f}')
        
        ax2.set_title(f'Histogramme - {var}', fontweight='bold')
        ax2.set_xlabel(var, fontweight='bold')
        ax2.set_ylabel('Fréquence', fontweight='bold')
        ax2.grid(True, alpha=0.3)
        ax2.legend()
        
        plt.tight_layout()
        plt.savefig(f'{var}_analysis.png', dpi=300, bbox_inches='tight')
        plt.show()

# =================== STATISTIQUES DESCRIPTIVES ===================
def print_statistics():
    """Afficher les statistiques descriptives pour chaque variable par mois"""
    
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    
    print("=== STATISTIQUES DESCRIPTIVES PAR MOIS ===\n")
    
    for var in variables:
        print(f"\n--- {var} ---")
        
        # Créer un tableau avec les noms de mois
        stats = df.groupby('Month')[var].describe()
        stats.index = get_month_labels(stats.index)
        print(stats.round(3))
        
        # Test de normalité et outliers
        print(f"\nValeurs aberrantes pour {var}:")
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
        print(f"Nombre d'outliers: {len(outliers)}")
        
        # Statistiques par mois
        print(f"Moyennes mensuelles:")
        monthly_means = df.groupby('Month')[var].mean()
        for month, mean_val in monthly_means.items():
            print(f"  {month_names[month]}: {mean_val:.3f}")
        
        print("-" * 60)

# =================== FONCTION PRINCIPALE ===================
def main():
    """Fonction principale pour exécuter toutes les analyses"""
    
    print("Chargement des données...")
    print(f"Dimensions du dataset: {df.shape}")
    print(f"Colonnes disponibles: {list(df.columns)}")
    print(f"Mois disponibles: {get_month_labels(sorted(df['Month'].unique()))}")
    print(f"Années disponibles: {sorted(df['Year'].unique())}")
    
    # Vérifier les données manquantes
    print("\nDonnées manquantes par variable:")
    missing_data = df.isnull().sum()
    for col, missing in missing_data.items():
        if missing > 0:
            print(f"{col}: {missing} ({missing/len(df)*100:.1f}%)")
    
    # Créer les visualisations
    print("\nCréation des box plots...")
    create_boxplots()
    
    print("Création des histogrammes...")
    create_histograms()
    
    print("Création des analyses mensuelles détaillées...")
    create_monthly_analysis()
    
    # Afficher les statistiques
    print_statistics()
    
    print("\nAnalyse terminée! Les graphiques ont été sauvegardés.")

# =================== FONCTIONS UTILITAIRES SUPPLÉMENTAIRES ===================

def create_correlation_matrix():
    """Créer une matrice de corrélation entre les variables"""
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    
    plt.figure(figsize=(10, 8))
    correlation_matrix = df[variables].corr()
    
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                square=True, linewidths=0.5, cbar_kws={"shrink": .8})
    
    plt.title('Matrice de corrélation des variables océanographiques', 
              fontweight='bold', fontsize=14)
    plt.tight_layout()
    plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

def create_temporal_trends():
    """Créer des graphiques de tendances temporelles"""
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    
    # Moyennes mensuelles
    monthly_means = df.groupby('Month')[variables].mean()
    
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    fig.suptitle('Tendances temporelles des variables océanographiques', 
                 fontsize=16, fontweight='bold')
    axes_flat = axes.flatten()
    
    for i, var in enumerate(variables):
        ax = axes_flat[i]
        month_labels = get_month_labels(monthly_means.index)
        ax.plot(range(len(monthly_means.index)), monthly_means[var], 
                'o-', linewidth=2, markersize=6, color='steelblue')
        ax.set_title(f'{chr(65+i)}) {var}', fontweight='bold', loc='left')
        ax.set_xlabel('Mois', fontweight='bold')
        ax.set_ylabel(var, fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.set_xticks(range(len(monthly_means.index)))
        ax.set_xticklabels(month_labels, rotation=45)
        
        # Ajouter une ligne de tendance
        z = np.polyfit(range(len(monthly_means.index)), monthly_means[var], 1)
        p = np.poly1d(z)
        ax.plot(range(len(monthly_means.index)), p(range(len(monthly_means.index))), 
                "r--", alpha=0.6, linewidth=1)
    
    plt.tight_layout()
    plt.savefig('tendances_temporelles.png', dpi=300, bbox_inches='tight')
    plt.show()

def create_seasonal_comparison():
    """Créer une comparaison saisonnière"""
    variables = ['EKE', 'EA1', 'Radius1', 'SSH1', 'RNL1', 'EA2', 'Radius2', 'SSH2', 'RNL2']
    
    # Définir les saisons
    seasons = {
        'Hiver': [12, 1, 2],
        'Printemps': [3, 4, 5],
        'Été': [6, 7, 8],
        'Automne': [9, 10, 11]
    }
    
    # Ajouter une colonne saison
    df_seasons = df.copy()
    df_seasons['Season'] = df_seasons['Month'].map(
        lambda x: next((season for season, months in seasons.items() if x in months), 'Autre')
    )
    
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    fig.suptitle('Comparaison saisonnière des variables océanographiques', 
                 fontsize=16, fontweight='bold')
    axes_flat = axes.flatten()
    
    for i, var in enumerate(variables):
        ax = axes_flat[i]
        
        # Box plot par saison
        season_data = []
        season_labels = []
        for season in ['Hiver', 'Printemps', 'Été', 'Automne']:
            data = df_seasons[df_seasons['Season'] == season][var].dropna()
            if len(data) > 0:
                season_data.append(data)
                season_labels.append(season)
        
        if season_data:
            bp = ax.boxplot(season_data, labels=season_labels, patch_artist=True)
            
            # Couleurs saisonnières
            season_colors = ['lightblue', 'lightgreen', 'yellow', 'orange']
            for patch, color in zip(bp['boxes'], season_colors[:len(season_data)]):
                patch.set_facecolor(color)
                patch.set_alpha(0.7)
            
            # Outliers en rouge
            for flier in bp['fliers']:
                flier.set(marker='o', color='red', alpha=0.6, markersize=3)
        
        ax.set_title(f'{chr(65+i)}) {var}', fontweight='bold', loc='left')
        ax.set_xlabel('Saison', fontweight='bold')
        ax.set_ylabel(var, fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.savefig('comparaison_saisonniere.png', dpi=300, bbox_inches='tight')
    plt.show()

# Exécuter l'analyse principale
if __name__ == "__main__":
    main()
    
    # Décommenter les lignes suivantes pour exécuter les analyses supplémentaires
    # print("\nCréation de la matrice de corrélation...")
    # create_correlation_matrix()
    # 
    # print("Création des tendances temporelles...")
    # create_temporal_trends()
    # 
    # print("Création de la comparaison saisonnière...")
    # create_seasonal_comparison()
```
